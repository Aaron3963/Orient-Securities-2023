```python
import pandas as pd
import numpy as np
import talib as ta

import seaborn as sns
import matplotlib.pyplot as plt

PATH = '123181_train_withF_cutMax.csv'

# df = pd.read_csv(PATH_READ, index_col='Timestamp')
df = pd.read_csv(PATH)

print(df.columns)
```

See trend

```python
sns.lineplot(data=df['Target_Close_1min'], color='r', alpha=0.5)
```

Clipping abnormal values

```python
df['Target_Avg_1min'] = df['Target_Avg_1min'].clip(lower=-0.02, upper=0.02)
df['Target_Avg_5min'] = df['Target_Avg_5min'].clip(lower=-0.02, upper=0.025)

df['Target_Close_1min'] = df['Target_Close_1min'].clip(lower=-0.025, upper=0.025)
df['Target_Close_5min'] = df['Target_Close_5min'].clip(lower=-0.025, upper=0.025)
```

```python
sns.lineplot(data=df['Target_Close_1min'], color='r', alpha=0.5)
```

```python
df.to_csv(PATH, index= False)
```
